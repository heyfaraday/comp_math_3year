```python
import numpy as np
```

```python
a = 0.
b = 1.
h = 0.02

x = np.arange(a, b + h, h)
y = 1. + np.sin(x)
n = np.arange(0, x.size, 1)
```

```python
number_of_iterations = 1000
```

```python
def f(x, y):
    return 2. * x + np.cos(y)
```

```python
for i in range(0, number_of_iterations):
    
    k1 = f(x, y)
    k2 = f(x + h / 2., y + h / 2. * k1)
    k3 = f(x + h / 2., y + h / 2. * k2)
    k4 = f(x + h, y + h * k3)
    
    y[1:-1] = y[0:-2] + (h * (k1 / 6. + k2 / 3. + k3 / 3. + k4 / 6.))[:-2]
```
