```python
import numpy as np
```

```python
tau = 0.05
beta = 0.91
```

```python
number_of_iterations = 200
```

```python
def grad_f(x):
    return np.array([3. * x[0] * x[0] - 3 * x[1], 3. * x[1] * x[1] - 3 * x[0]])
def f(x):
    return x[0]**3 + x[1]**3 - 3 * x[0] * x[1]
```

```python
import matplotlib.pyplot as plt

x_0 = np.array([3, 3])
v_0 = np.array([0., 0.])

x = np.arange(-5, 5, 0.05)
y = np.arange(-5, 5, 0.05)
xgrid, ygrid = np.meshgrid(x, y)

zgrid = f(np.array([xgrid, ygrid]))

plt.figure(figsize=(15, 15))
plt.contour(xgrid, ygrid, zgrid, 50)

for i in range(0, number_of_iterations):
    plt.scatter([x_0[0]], [x_0[1]], marker='o', s=20, c='r')
    v_dash = v_0 - grad_f(x_0) * tau
    v_0 = beta * v_dash
    x_0 = x_0 + v_0 * tau
    
#     print(x_0)

plt.show()
```
